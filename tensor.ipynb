<h3 style ="color:white;background-color:blue;padding : 10px;text-align:center"><b>IDENTIFICANDO AS LETRAS NAS IMAGENS</b></h3>

```python
padding = 10
imagem = cv.imread("tela.png")
positions = pytesseract.image_to_boxes("tela.png").split("\n")
index = 0
img_cv = cv.imread("tela.png")
positions.pop()
img_flip = cv.flip(img_cv,0)
for i in positions:
    draw_positions = positions[index].split()
    _, x1, y1, x2, y2, _ = draw_positions
    cv.rectangle(img_flip,(int(x1) - padding, int(y1)), (int(x2) + padding, int(y2)), (255, 0, 0), 2, cv.LINE_AA)
    index = index + 1

plt.figure(figsize=[20,5])    
plt.subplot(121);plt.title("Imagem da tela");plt.imshow(imagem)
plt.subplot(122);plt.title("Letras identificadas");plt.imshow(img)
img_flip = cv.flip(img_flip,0)
plt.imshow(img_flip)
```

<h3 style ="color:white;background-color:blue;padding : 10px;text-align:center"><b>IDENTIFICANDO AS PALAVRAS NAS IMAGENS</b></h3>

```python
index = 0
padding = 10
imagem = cv.imread("tela.png")

positions = pytesseract.image_to_boxes("tela.png").split("\n")
positions.pop()
letras = pytesseract.image_to_string("tela.png")
letras = letras.replace("\n\n",' ')
letras = letras.split()
img_cv = cv.imread("tela.png")
img_flip = cv.flip(img_cv,0)

for i in letras:
    draw_positions = positions[index].split()
    _, x1, y1, x2, y2, _ = draw_positions
    cv.rectangle(img_flip,(int(x1),int(y1)),(int(x2),int(y2)), (255, 0, 0), 2, cv.LINE_AA)
    index = index + 1

plt.figure(figsize=[20,5])    
plt.subplot(121);plt.title("Imagem da tela");plt.imshow(imagem)
plt.subplot(122);plt.title("Letras identificadas");plt.imshow(img)
img_flip = cv.flip(img_flip,0)
plt.imshow(img_flip)
```
